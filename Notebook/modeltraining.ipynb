```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\Admin\\Desktop\\AnamolyDetection\\Notebook\\data\\creditcard_data.csv')
```

```python
df.head()
```

```python
X = df.drop(columns=['Class'])  
y = df['Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn.metrics import classification_report, roc_auc_score
import xgboost as xgb

# Assuming you already have X (features) and y (target) defined



# Create XGBoost classifier instance
xgb_classifier = xgb.XGBClassifier()

# Fit XGBoost classifier on training data
xgb_classifier.fit(X_train, y_train)

# Make predictions on test data
predictions = xgb_classifier.predict(X_test)

# Compute classification report
report = classification_report(y_test, predictions)

# Compute ROC AUC score
auc_roc_score = roc_auc_score(y_test, predictions)

print("Classification Report:")
print(report)
print("ROC AUC Score:", auc_roc_score)
```

```python
# Define best parameters obtained from parameter tuning
best_parameters = {'colsample_bytree': 0.8, 'gamma': 0, 'learning_rate': 0.3, 'max_depth': 7, 'n_estimators': 100, 'subsample': 0.8, }

# Now you can proceed with the rest of your code
```

```python
import pickle
from sklearn.metrics import classification_report, roc_auc_score
import xgboost as xgb

# Assuming you already have X_train, X_test, y_train, y_test defined and the best parameters

# Fit XGBoost classifier with the best parameters
best_xgb_classifier = xgb.XGBClassifier(**best_parameters)
best_xgb_classifier.fit(X_train, y_train)

# Make predictions on test data
predictions = best_xgb_classifier.predict(X_test)

# Compute classification report
report = classification_report(y_test, predictions)

# Compute ROC AUC score
auc_roc_score = roc_auc_score(y_test, predictions)

print("Classification Report:")
print(report)
print("ROC AUC Score:", auc_roc_score)

# Save the trained model with pickle
with open('xgb_model.pkl', 'wb') as f:
    pickle.dump(best_xgb_classifier, f)

# Feature importances
feature_importances = best_xgb_classifier.feature_importances_
print("Feature Importances:")
for i, importance in enumerate(feature_importances):
    print("Feature {}: {:.4f}".format(i + 1, importance))
```
